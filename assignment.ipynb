```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Load the data
wainaina = pd.read_csv('/content/Accidents.csv')
# Extract the independent variable (X) and dependent variable (y)
X = wainaina[['Number_of_Vehicles']]
y = wainaina['Accident_Severity']
# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X, y)
# Plot the data points
wainaina.plot(x='Number_of_Vehicles', y='Accident_Severity', style='o', label='Data Points')
# Overlay the regression line
plt.plot(X, regressor.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Number_of_Vehicles')
plt.ylabel('Accident_Severity')
plt.legend()
plt.show()
```
